```python
from matplotlib import pyplot as plt
import random 
import math
%matplotlib inline
from operator import itemgetter 
plt.rcParams['figure.figsize'] = (16, 8)
plt.rcParams.update({'font.size': 14})
```

```python
FILENAME = 'data/alice.txt'
```

```python
def getLines(filename):
    with open(filename) as datapointer:
        allLines = datapointer.readlines()
    return allLines
```

```python
def countWords(filename):
    wordCount = {}
    lines = getLines(filename)
    for aline in lines:
        for word in aline.split():
            if word in wordCount:
                wordCount[word]+=1
            else:
                wordCount[word] = 0 
    return wordCount
```

```python
wordCount = countWords(FILENAME)
```

```python
sortedCounts = sorted(wordCount.items(), key = itemgetter(1),reverse = True)
```

```python
top20words = sortedCounts[:20]
words = []
counts = []
for word,count in top20words:
    words.append(word)
    counts.append(count)
```

```python
plt.bar(range(len(words)), counts, color='green',width=0.9)
plt.xticks(range(len(words)),words,rotation=65)
plt.xlabel('Vocabulary')
plt.ylabel('Counts')
plt.title('TOP 20 MOST COMMON WORDS')
plt.show()
```
