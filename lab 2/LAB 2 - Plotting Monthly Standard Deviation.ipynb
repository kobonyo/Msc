```python
import random 
import math 
%matplotlib inline
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams.update({'font.size': 14})
```

```python
months = ['JANUARY','FEBRUARY','MARCH','APRIL',\
          'MAY','JUNE','JULY','AUGUST','SEPTEMBER',\
          'OCTOBER','NOVEMBER','DECEMBER'
         ]
```

```python
# generating the monthly tempreatures
monthly_tempratures = {}
for val in months:
    monthly_tempratures[val] = random.sample(range(50),30)
```

```python
#function to calculate sum or average

def calculateSumAvg(values,avg=False):
    sum_ = 0
    for i in range(len(values)):
        sum_+= values[i]
    if not avg:
        return sum_
    return sum_/len(values)
```

```python
#fucnction to calculate std
def getStd(monthlyVals):
    m_std = {}
    for amonth in months:
        variance = []
        mean = calculateSumAvg(monthlyVals[amonth],avg=True)
        for temp  in monthlyVals[amonth]:
            variance.append((mean-temp)**2)
        m_std[amonth] = math.sqrt(calculateSumAvg(variance)/(30-1))
        variance = []
    return m_std
```

```python
# get monthly std
monthlystd = getStd(monthly_tempratures)
# scatter plot for monthly std
x = list(monthlystd.values())
y = list(monthlystd.keys())
plt.scatter(list(range(1,len(y)+1)),x,s=70)
plt.xticks(range(len(y)),y,rotation=65)
plt.ylabel('Standard Deviation')
plt.xlabel('Month of the year')
plt.title('Monthly Standard Deviation')
plt.savefig("std.png")
plt.show()
```
